```python
import pandas as pd
import requests
import time

df = pd.DataFrame()

for i in range(1, 52):
    url = 'https://api.themoviedb.org/3/movie/upcoming?api_key=2593f6e6803c56dea3a0f4152434e7c2&language=en-US&'.format(i)
    
    success = False
    
    while not success:
        try:
            result = requests.get(url, timeout=10)
            data = result.json()
            
            # safety check
            if 'results' not in data:
                print(f"⚠️ No data on page {i}")
                break
            
            temp_df = pd.DataFrame(data['results'])[
                ['id','title','overview','release_date','popularity','vote_average','vote_count']
            ]
            
            df = pd.concat([df, temp_df], ignore_index=True)
            
            print(f"✅ Page {i} done")
            
            success = True
            time.sleep(1)   # 🔥 prevent blocking
        
        except Exception as e:
            print(f"❌ Error on page {i}: {e}")
            print("Retrying...")
            time.sleep(3)

print("Final shape:", df.shape)
```
